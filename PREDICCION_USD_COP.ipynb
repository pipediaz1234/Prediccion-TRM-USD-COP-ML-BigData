```python
"""
============================================================
SISTEMA AVANZADO DE PREDICCI√ìN USD/COP (TRM)
Banco de la Rep√∫blica de Colombia
============================================================
Autor: Andres Felipe Diaz
Fecha: 19 Octubre 2025
============================================================
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from google.colab import files
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# ============================================================
# CONFIGURACI√ìN VISUAL PROFESIONAL
# ============================================================

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

COLORES = {
    'principal': '#1A5490',
    'secundario': '#FF6B35',
    'exito': '#2EA043',
    'alerta': '#DA3633',
    'info': '#8B5CF6',
    'neutral': '#6B7280',
    'acento': '#F59E0B'
}

plt.rcParams.update({
    'figure.figsize': (14, 6),
    'font.size': 10,
    'axes.titlesize': 13,
    'axes.labelsize': 11,
    'axes.titleweight': 'bold',
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'legend.fontsize': 9,
    'figure.titlesize': 16
})


class PredictorTRMAvanzado:
    """
    Sistema profesional de predicci√≥n TRM con m√∫ltiples modelos ML
    y an√°lisis t√©cnico-financiero avanzado
    """

    def __init__(self):
        self.df = None
        self.modelo_final = None
        self.scaler = StandardScaler()
        self.metricas = {}
        self.df_futuro = None
        self.modelos_comparados = {}

    def _print_seccion(self, titulo, emoji="üìä"):
        """Imprime secci√≥n formateada"""
        print("\n" + "="*70)
        print(f"{emoji} {titulo}")
        print("="*70)

    def _print_subseccion(self, texto):
        """Imprime subsecci√≥n"""
        print(f"\n{texto}")
        print("-" * 70)

    def cargar_datos(self, filepath=None):
        """Carga y limpia datos del Banco de la Rep√∫blica"""
        self._print_seccion("M√ìDULO 1: CARGA Y PREPARACI√ìN DE DATOS", "üì•")

        if filepath is None:
            print("\nüìÅ Sube el archivo CSV del Banco de la Rep√∫blica:")
            uploaded = files.upload()
            filepath = list(uploaded.keys())[0]

        # Intentar m√∫ltiples encodings
        encodings = ['latin-1', 'utf-8', 'iso-8859-1', 'cp1252']
        df = None

        for enc in encodings:
            try:
                df = pd.read_csv(filepath, sep=";", encoding=enc)
                print(f"‚úÖ Archivo cargado con encoding: {enc}")
                break
            except:
                continue

        if df is None:
            raise ValueError("‚ùå No se pudo leer el archivo con ning√∫n encoding")

        # Limpieza de columnas
        df.columns = [c.strip().replace('"', '').replace("'", "") for c in df.columns]

        # Identificar columnas autom√°ticamente
        fecha_col = None
        valor_col = None

        for col in df.columns:
            col_lower = col.lower()
            if any(x in col_lower for x in ['periodo', 'fecha', 'date', 'vigencia']):
                fecha_col = col
            if any(x in col_lower for x in ['trm', 'tasa', 'valor', 'cambio']):
                valor_col = col

        if not fecha_col or not valor_col:
            raise ValueError("‚ùå No se identificaron las columnas necesarias")

        print(f"\nüìã Columnas detectadas:")
        print(f"   ‚Ä¢ Fecha: '{fecha_col}'")
        print(f"   ‚Ä¢ TRM: '{valor_col}'")

        # Seleccionar y renombrar
        df = df[[fecha_col, valor_col]].copy()
        df.columns = ['Fecha', 'USD_COP']

        # Limpieza robusta
        df['Fecha'] = df['Fecha'].astype(str).str.strip()
        df['USD_COP'] = (df['USD_COP'].astype(str).str.strip()
                        .str.replace('.', '', regex=False)
                        .str.replace(',', '.', regex=False)
                        .str.replace('$', '', regex=False)
                        .str.replace(' ', '', regex=False))

        # Conversi√≥n de tipos
        df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce', dayfirst=True)
        df['USD_COP'] = pd.to_numeric(df['USD_COP'], errors='coerce')

        # Limpieza final
        registros_inicial = len(df)
        df = df.dropna(subset=['Fecha', 'USD_COP'])
        df = df.drop_duplicates(subset=['Fecha'])
        df = df.sort_values('Fecha').reset_index(drop=True)
        registros_final = len(df)

        # Validaci√≥n de calidad
        if registros_final < 100:
            raise ValueError(f"‚ùå Datos insuficientes: solo {registros_final} registros v√°lidos")

        self._print_subseccion("üìä Resumen de Carga")
        print(f"   ‚Ä¢ Registros cargados: {registros_inicial:,}")
        print(f"   ‚Ä¢ Registros v√°lidos: {registros_final:,}")
        print(f"   ‚Ä¢ Registros eliminados: {registros_inicial - registros_final:,}")
        print(f"   ‚Ä¢ Rango: {df['Fecha'].min().date()} ‚Üí {df['Fecha'].max().date()}")
        print(f"   ‚Ä¢ D√≠as totales: {(df['Fecha'].max() - df['Fecha'].min()).days:,}")
        print(f"   ‚Ä¢ A√±os de historia: {(df['Fecha'].max() - df['Fecha'].min()).days / 365:.1f}")

        self.df = df
        return df

    def analisis_exploratorio(self):
        """An√°lisis exploratorio completo con visualizaciones profesionales"""
        self._print_seccion("M√ìDULO 2: AN√ÅLISIS EXPLORATORIO AVANZADO", "üìä")

        df = self.df.copy()

        # C√°lculos estad√≠sticos
        df['Retorno_Diario'] = df['USD_COP'].pct_change() * 100
        df['Retorno_30d'] = df['USD_COP'].pct_change(30) * 100
        df['Volatilidad_30d'] = df['Retorno_Diario'].rolling(30).std()
        df['A√±o'] = df['Fecha'].dt.year
        df['Mes'] = df['Fecha'].dt.month

        # Estad√≠sticas descriptivas
        self._print_subseccion("üí∞ Estad√≠sticas Descriptivas")
        stats_desc = df['USD_COP'].describe()

        print(f"\n{'M√©trica':<25} {'Valor':>15}")
        print("-" * 45)
        print(f"{'M√≠nimo hist√≥rico':<25} ${stats_desc['min']:>14,.2f}")
        print(f"{'M√°ximo hist√≥rico':<25} ${stats_desc['max']:>14,.2f}")
        print(f"{'Promedio':<25} ${stats_desc['mean']:>14,.2f}")
        print(f"{'Mediana':<25} ${stats_desc['50%']:>14,.2f}")
        print(f"{'Desviaci√≥n est√°ndar':<25} ${stats_desc['std']:>14,.2f}")
        print(f"{'Rango':<25} ${stats_desc['max'] - stats_desc['min']:>14,.2f}")

        # Volatilidad
        print(f"\n{'Volatilidad diaria (avg)':<25} {df['Retorno_Diario'].std():>14.4f}%")
        print(f"{'Retorno diario (avg)':<25} {df['Retorno_Diario'].mean():>14.4f}%")
        print(f"{'Mejor d√≠a (ganancia)':<25} {df['Retorno_Diario'].max():>14.2f}%")
        print(f"{'Peor d√≠a (p√©rdida)':<25} {df['Retorno_Diario'].min():>14.2f}%")

        # Tendencias por a√±o
        self._print_subseccion("üìÖ An√°lisis por A√±o (√∫ltimos 10 a√±os)")
        a√±os_recientes = sorted(df['A√±o'].unique())[-10:]
        resumen_anual = df[df['A√±o'].isin(a√±os_recientes)].groupby('A√±o').agg({
            'USD_COP': ['mean', 'min', 'max', 'std'],
            'Retorno_Diario': 'std'
        }).round(2)
        print(resumen_anual.to_string())

        # Dashboard visual completo
        fig = plt.figure(figsize=(18, 12))
        gs = fig.add_gridspec(3, 3, hspace=0.35, wspace=0.3)

        # 1. Serie temporal principal
        ax1 = fig.add_subplot(gs[0, :])
        ax1.plot(df['Fecha'], df['USD_COP'], color=COLORES['principal'],
                linewidth=1.5, alpha=0.8, label='TRM')
        ax1.fill_between(df['Fecha'], df['USD_COP'], alpha=0.2, color=COLORES['principal'])

        # Marcar m√≠nimo y m√°ximo hist√≥rico
        idx_min = df['USD_COP'].idxmin()
        idx_max = df['USD_COP'].idxmax()
        ax1.scatter(df.loc[idx_min, 'Fecha'], df.loc[idx_min, 'USD_COP'],
                   color=COLORES['exito'], s=100, zorder=5, marker='v', label='M√≠nimo hist√≥rico')
        ax1.scatter(df.loc[idx_max, 'Fecha'], df.loc[idx_max, 'USD_COP'],
                   color=COLORES['alerta'], s=100, zorder=5, marker='^', label='M√°ximo hist√≥rico')

        ax1.set_title('üìà Evoluci√≥n Hist√≥rica del Tipo de Cambio USD/COP (TRM)',
                     fontsize=14, pad=15)
        ax1.set_xlabel('Fecha')
        ax1.set_ylabel('TRM (COP/USD)')
        ax1.legend(loc='best')
        ax1.grid(True, alpha=0.3)
        ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

        # 2. Distribuci√≥n de precios
        ax2 = fig.add_subplot(gs[1, 0])
        ax2.hist(df['USD_COP'], bins=60, color=COLORES['info'],
                alpha=0.7, edgecolor='black', density=True)
        ax2.axvline(df['USD_COP'].mean(), color=COLORES['alerta'],
                   linestyle='--', linewidth=2, label=f'Media: ${df["USD_COP"].mean():,.0f}')
        ax2.axvline(df['USD_COP'].median(), color=COLORES['exito'],
                   linestyle='--', linewidth=2, label=f'Mediana: ${df["USD_COP"].median():,.0f}')
        ax2.set_title('üìä Distribuci√≥n de Valores TRM', pad=10)
        ax2.set_xlabel('TRM (COP/USD)')
        ax2.set_ylabel('Densidad')
        ax2.legend()
        ax2.grid(True, alpha=0.3, axis='y')

        # 3. Retornos diarios
        ax3 = fig.add_subplot(gs[1, 1])
        colores_retornos = [COLORES['exito'] if x > 0 else COLORES['alerta']
                           for x in df['Retorno_Diario']]
        ax3.scatter(df['Fecha'], df['Retorno_Diario'], c=colores_retornos,
                   alpha=0.4, s=5)
        ax3.axhline(0, color='black', linestyle='-', linewidth=1)
        ax3.set_title('üìâ Retornos Diarios (%)', pad=10)
        ax3.set_xlabel('Fecha')
        ax3.set_ylabel('Cambio %')
        ax3.grid(True, alpha=0.3)
        plt.setp(ax3.xaxis.get_majorticklabels(), rotation=45)

        # 4. Volatilidad rodante
        ax4 = fig.add_subplot(gs[1, 2])
        ax4.plot(df['Fecha'], df['Volatilidad_30d'], color=COLORES['secundario'],
                linewidth=1.5)
        ax4.fill_between(df['Fecha'], df['Volatilidad_30d'], alpha=0.3,
                        color=COLORES['secundario'])
        ax4.set_title(' Volatilidad 30 d√≠as', pad=10)
        ax4.set_xlabel('Fecha')
        ax4.set_ylabel('Volatilidad %')
        ax4.grid(True, alpha=0.3)
        plt.setp(ax4.xaxis.get_majorticklabels(), rotation=45)

        # 5. √öltimos 180 d√≠as
        ax5 = fig.add_subplot(gs[2, 0])
        df_reciente = df.tail(180)
        ax5.plot(df_reciente['Fecha'], df_reciente['USD_COP'],
                color=COLORES['acento'], linewidth=2, marker='o', markersize=3)
        ax5.set_title(' √öltimos 6 Meses', pad=10)
        ax5.set_xlabel('Fecha')
        ax5.set_ylabel('TRM (COP/USD)')
        ax5.grid(True, alpha=0.3)
        plt.setp(ax5.xaxis.get_majorticklabels(), rotation=45)

        # 6. Boxplot por a√±o
        ax6 = fig.add_subplot(gs[2, 1])
        df_box = df[df['A√±o'].isin(a√±os_recientes)]
        bp = ax6.boxplot([df_box[df_box['A√±o']==a√±o]['USD_COP'].values
                          for a√±o in a√±os_recientes],
                         labels=a√±os_recientes, patch_artist=True)
        for patch in bp['boxes']:
            patch.set_facecolor(COLORES['info'])
            patch.set_alpha(0.6)
        ax6.set_title('Distribuci√≥n Anual (10 a√±os)', pad=10)
        ax6.set_xlabel('A√±o')
        ax6.set_ylabel('TRM (COP/USD)')
        ax6.grid(True, alpha=0.3, axis='y')
        plt.setp(ax6.xaxis.get_majorticklabels(), rotation=45)

        # 7. Heatmap estacionalidad
        ax7 = fig.add_subplot(gs[2, 2])
        pivot_data = df.groupby(['A√±o', 'Mes'])['USD_COP'].mean().reset_index()
        pivot_table = pivot_data.pivot(index='Mes', columns='A√±o', values='USD_COP')
        pivot_recent = pivot_table[a√±os_recientes[-5:]]

        sns.heatmap(pivot_recent, ax=ax7, cmap='RdYlGn_r',
                   fmt='.0f', cbar_kws={'label': 'TRM promedio'},
                   linewidths=0.5)
        ax7.set_title(' Estacionalidad (5 a√±os)', pad=10)
        ax7.set_xlabel('A√±o')
        ax7.set_ylabel('Mes')

        plt.tight_layout()
        plt.show()

        self.df = df
        return df

    def ingenieria_caracteristicas(self):
        """Feature engineering avanzado con indicadores t√©cnicos"""
        self._print_seccion("M√ìDULO 3: INGENIER√çA DE CARACTER√çSTICAS AVANZADA", "üîß")

        df = self.df.copy()
        registros_inicial = len(df)

        print("\n‚öôÔ∏è  Creando variables predictoras...")

        # Variables temporales
        df['Dia'] = np.arange(len(df))
        df['Mes'] = df['Fecha'].dt.month
        df['Trimestre'] = df['Fecha'].dt.quarter
        df['DiaSemana'] = df['Fecha'].dt.dayofweek
        df['DiaAnio'] = df['Fecha'].dt.dayofyear
        df['Semana'] = df['Fecha'].dt.isocalendar().week

        # Lags m√∫ltiples
        lags = [1, 2, 3, 5, 7, 10, 14, 20, 21, 30, 60, 90]
        for lag in lags:
            df[f'Lag_{lag}'] = df['USD_COP'].shift(lag)

        # Medias m√≥viles (SMA)
        ventanas_ma = [5, 10, 20, 30, 50, 100, 200]
        for ventana in ventanas_ma:
            df[f'SMA_{ventana}'] = df['USD_COP'].rolling(window=ventana).mean()

        # Medias m√≥viles exponenciales (EMA)
        ventanas_ema = [12, 26, 50]
        for ventana in ventanas_ema:
            df[f'EMA_{ventana}'] = df['USD_COP'].ewm(span=ventana, adjust=False).mean()

        # Volatilidad (desviaci√≥n est√°ndar rodante)
        ventanas_std = [5, 10, 20, 30]
        for ventana in ventanas_std:
            df[f'Volatilidad_{ventana}'] = df['USD_COP'].rolling(window=ventana).std()

        # Bandas de Bollinger
        df['BB_Media'] = df['SMA_20']
        df['BB_Superior'] = df['BB_Media'] + (2 * df['Volatilidad_20'])
        df['BB_Inferior'] = df['BB_Media'] - (2 * df['Volatilidad_20'])
        df['BB_Ancho'] = df['BB_Superior'] - df['BB_Inferior']
        df['BB_Posicion'] = (df['USD_COP'] - df['BB_Inferior']) / df['BB_Ancho']

        # Rate of Change (ROC)
        periodos_roc = [5, 10, 20, 30]
        for periodo in periodos_roc:
            df[f'ROC_{periodo}'] = ((df['USD_COP'] - df[f'Lag_{periodo}']) / df[f'Lag_{periodo}']) * 100

        # RSI (Relative Strength Index)
        def calcular_rsi(data, periodo=14):
            delta = data.diff()
            ganancia = delta.where(delta > 0, 0).rolling(window=periodo).mean()
            perdida = -delta.where(delta < 0, 0).rolling(window=periodo).mean()
            rs = ganancia / perdida
            return 100 - (100 / (1 + rs))

        df['RSI_14'] = calcular_rsi(df['USD_COP'], 14)

        # MACD (Moving Average Convergence Divergence)
        df['MACD'] = df['EMA_12'] - df['EMA_26']
        df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
        df['MACD_Histogram'] = df['MACD'] - df['MACD_Signal']

        # Momentum
        df['Momentum_10'] = df['USD_COP'] - df['Lag_10']
        df['Momentum_30'] = df['USD_COP'] - df['Lag_30']

        # Rango verdadero (True Range) - versi√≥n simplificada
        df['TR'] = df['USD_COP'].rolling(window=2).apply(lambda x: abs(x.iloc[-1] - x.iloc[0]))
        df['ATR_14'] = df['TR'].rolling(window=14).mean()

        # Cruce de medias m√≥viles
        df['MA_Cross_50_200'] = np.where(df['SMA_50'] > df['SMA_200'], 1, 0)

        # Distancia desde medias m√≥viles
        df['Dist_SMA_20'] = ((df['USD_COP'] - df['SMA_20']) / df['SMA_20']) * 100
        df['Dist_SMA_50'] = ((df['USD_COP'] - df['SMA_50']) / df['SMA_50']) * 100

        # Variables de tendencia
        df['Tendencia_5d'] = np.where(df['USD_COP'] > df['Lag_5'], 1, -1)
        df['Tendencia_10d'] = np.where(df['USD_COP'] > df['Lag_10'], 1, -1)
        df['Tendencia_20d'] = np.where(df['USD_COP'] > df['Lag_20'], 1, -1)

        # Eliminar NaN
        df = df.dropna()
        registros_final = len(df)

        self._print_subseccion("‚úÖ Resumen de Variables Creadas")
        print(f"   ‚Ä¢ Variables temporales: 6")
        print(f"   ‚Ä¢ Lags: 12")
        print(f"   ‚Ä¢ Medias m√≥viles simples: {len(ventanas_ma)}")
        print(f"   ‚Ä¢ Medias m√≥viles exponenciales: {len(ventanas_ema)}")
        print(f"   ‚Ä¢ Indicadores de volatilidad: {len(ventanas_std)}")
        print(f"   ‚Ä¢ Bandas de Bollinger: 5")
        print(f"   ‚Ä¢ ROC (Rate of Change): {len(periodos_roc)}")
        print(f"   ‚Ä¢ Indicadores t√©cnicos: 10+")
        print(f"   ‚Ä¢ Total de columnas: {len(df.columns)}")
        print(f"   ‚Ä¢ Registros v√°lidos: {registros_final:,}")
        print(f"   ‚Ä¢ Registros eliminados: {registros_inicial - registros_final:,}")

        self.df = df
        return df

    def entrenar_modelos(self, test_size=0.2):
        """Entrena y compara m√∫ltiples modelos de ML"""
        self._print_seccion("M√ìDULO 4: ENTRENAMIENTO Y COMPARACI√ìN DE MODELOS", "ü§ñ")

        df = self.df

        # Seleccionar features
        features = [col for col in df.columns if col not in
                   ['Fecha', 'USD_COP', 'A√±o', 'Retorno_Diario', 'Retorno_30d',
                    'Volatilidad_30d', 'TR']]

        print(f"\n Features seleccionadas: {len(features)}")
        print(f"   Primeras 10: {features[:10]}")

        X = df[features]
        y = df['USD_COP']

        # Divisi√≥n temporal
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, shuffle=False
        )

        self._print_subseccion("üìä Divisi√≥n de Datos")
        print(f"   ‚Ä¢ Entrenamiento: {len(X_train):,} registros ({(1-test_size)*100:.0f}%)")
        print(f"   ‚Ä¢ Prueba: {len(X_test):,} registros ({test_size*100:.0f}%)")
        print(f"   ‚Ä¢ Fecha inicio entrenamiento: {df.loc[X_train.index[0], 'Fecha'].date()}")
        print(f"   ‚Ä¢ Fecha inicio prueba: {df.loc[X_test.index[0], 'Fecha'].date()}")

        # Normalizaci√≥n
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)

        # Definir modelos
        modelos = {
            'Regresi√≥n Lineal': LinearRegression(),
            'Ridge Regression': Ridge(alpha=1.0),
            'Lasso Regression': Lasso(alpha=0.1, max_iter=5000),
            'Random Forest': RandomForestRegressor(n_estimators=100, max_depth=15,
                                                   random_state=42, n_jobs=-1),
            'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, max_depth=5,
                                                           random_state=42)
        }

        print("\n Entrenando modelos...")
        resultados = {}

        for nombre, modelo in modelos.items():
            print(f"   ‚Ä¢ Entrenando {nombre}...", end=' ')

            # Entrenar
            modelo.fit(X_train_scaled, y_train)

            # Predicciones
            y_pred_train = modelo.predict(X_train_scaled)
            y_pred_test = modelo.predict(X_test_scaled)

            # M√©tricas
            rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
            rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
            mae_test = mean_absolute_error(y_test, y_pred_test)
            mape_test = mean_absolute_percentage_error(y_test, y_pred_test) * 100
            r2_train = r2_score(y_train, y_pred_train)
            r2_test = r2_score(y_test, y_pred_test)

            resultados[nombre] = {
                'modelo': modelo,
                'y_pred_train': y_pred_train,
                'y_pred_test': y_pred_test,
                'rmse_train': rmse_train,
                'rmse_test': rmse_test,
                'mae_test': mae_test,
                'mape_test': mape_test,
                'r2_train': r2_train,
                'r2_test': r2_test
            }

            print("‚úì")

        # Seleccionar mejor modelo
        mejor_modelo_nombre = min(resultados.items(), key=lambda x: x[1]['rmse_test'])[0]
        self.modelo_final = resultados[mejor_modelo_nombre]['modelo']
        self.metricas = resultados[mejor_modelo_nombre]
        self.modelos_comparados = resultados

        # Mostrar resultados
        self._print_subseccion("üìà RESULTADOS DE EVALUACI√ìN")

        print(f"\n{'Modelo':<20} {'RMSE Test':>12} {'MAE':>10} {'MAPE':>8} {'R¬≤ Test':>9} {'Status':>8}")
        print("-" * 75)

        for nombre, res in sorted(resultados.items(), key=lambda x: x[1]['rmse_test']):
            marcador = "‚≠ê MEJOR" if nombre == mejor_modelo_nombre else ""
            print(f"{nombre:<20} ${res['rmse_test']:>11,.2f} "
                  f"${res['mae_test']:>9,.2f} "
                  f"{res['mape_test']:>7.2f}% "
                  f"{res['r2_test']:>8.4f} "
                  f"{marcador:>8}")

        self._print_subseccion(" Interpretaci√≥n del Mejor Modelo")
        print(f"   ‚úÖ Modelo seleccionado: {mejor_modelo_nombre}")
        print(f"   ‚Ä¢ Poder explicativo (R¬≤): {self.metricas['r2_test']*100:.2f}%")
        print(f"   ‚Ä¢ Error promedio: ${self.metricas['mae_test']:,.2f} COP")
        print(f"   ‚Ä¢ Error porcentual: {self.metricas['mape_test']:.2f}%")
        print(f"   ‚Ä¢ RMSE: ${self.metricas['rmse_test']:,.2f}")

        # Importancia de features (si aplica)
        if hasattr(self.modelo_final, 'feature_importances_'):
            importancias = pd.DataFrame({
                'Feature': features,
                'Importancia': self.modelo_final.feature_importances_
            }).sort_values('Importancia', ascending=False).head(15)

            print("\nüîù Top 15 caracter√≠sticas m√°s importantes:")
            for idx, row in importancias.iterrows():
                barra = '‚ñà' * int(row['Importancia'] * 100)
                print(f"   {row['Feature']:<25} {barra} {row['Importancia']:.4f}")

        # Guardar datos para uso posterior
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
        self.X_train_scaled = X_train_scaled
        self.X_test_scaled = X_test_scaled
        self.features = features

        return self.modelo_final

    def visualizar_resultados(self):
        """Visualizaci√≥n profesional de resultados del modelo"""
        self._print_seccion("M√ìDULO 5: VISUALIZACI√ìN DE RESULTADOS", "üìä")

        df = self.df
        y_pred_test = self.modelo_final.predict(self.X_test_scaled)
        residuos = self.y_test - y_pred_test

        # Dashboard de resultados
        fig = plt.figure(figsize=(18, 12))
        gs = fig.add_gridspec(3, 3, hspace=0.35, wspace=0.3)

        # 1. Predicci√≥n vs Real (√∫ltimos 6 meses)
        ax1 = fig.add_subplot(gs[0, :2])
        fechas_test = df.loc[self.y_test.index, 'Fecha']
        n_mostrar = min(180, len(fechas_test))

        fechas_plot = fechas_test.iloc[-n_mostrar:]
        real_plot = self.y_test.iloc[-n_mostrar:].values
        pred_plot = y_pred_test[-n_mostrar:]

        ax1.plot(fechas_plot, real_plot, label='Valor Real',
                color=COLORES['principal'], linewidth=2.5, marker='o',
                markersize=4, alpha=0.8)
        ax1.plot(fechas_plot, pred_plot, label='Predicci√≥n del Modelo',
                color=COLORES['secundario'], linewidth=2.5, linestyle='--',
                marker='s', markersize=4, alpha=0.8)
        ax1.fill_between(fechas_plot, real_plot, pred_plot,
                        alpha=0.2, color=COLORES['neutral'])

        ax1.set_title('üìà Predicci√≥n vs Valor Real (√öltimos 6 meses)',
                     fontsize=14, pad=15)
        ax1.set_xlabel('Fecha', fontsize=11)
        ax1.set_ylabel('TRM (COP/USD)', fontsize=11)
        ax1.legend(loc='best', fontsize=10)
        ax1.grid(True, alpha=0.3)
        ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
        plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45)

        # 2. Comparaci√≥n de modelos (R¬≤)
        ax2 = fig.add_subplot(gs[0, 2])
        modelos_nombres = list(self.modelos_comparados.keys())
        r2_scores = [self.modelos_comparados[m]['r2_test'] for m in modelos_nombres]

        colores_barras = [COLORES['exito'] if r2 == max(r2_scores)
                         else COLORES['info'] for r2 in r2_scores]

        barras = ax2.barh(modelos_nombres, r2_scores, color=colores_barras, alpha=0.7)
        ax2.set_title('üèÜ Comparaci√≥n R¬≤ Score', pad=10)
        ax2.set_xlabel('R¬≤ Score')
        ax2.set_xlim([min(r2_scores)*0.95, 1.0])
        ax2.grid(True, alpha=0.3, axis='x')

        for i, (nombre, valor) in enumerate(zip(modelos_nombres, r2_scores)):
            ax2.text(valor + 0.005, i, f'{valor:.4f}',
                    va='center', fontsize=9, fontweight='bold')

        # 3. Scatter: Real vs Predicci√≥n
        ax3 = fig.add_subplot(gs[1, 0])
        ax3.scatter(self.y_test, y_pred_test, alpha=0.5, s=30,
                   c=COLORES['info'], edgecolors='black', linewidth=0.5)

        min_val = min(self.y_test.min(), y_pred_test.min())
        max_val = max(self.y_test.max(), y_pred_test.max())
        ax3.plot([min_val, max_val], [min_val, max_val],
                'r--', linewidth=2, label='Predicci√≥n Perfecta')

        ax3.set_title(f'üéØ Precisi√≥n (R¬≤={self.metricas["r2_test"]:.4f})', pad=10)
        ax3.set_xlabel('Valor Real (COP/USD)')
        ax3.set_ylabel('Valor Predicho (COP/USD)')
        ax3.legend(fontsize=9)
        ax3.grid(True, alpha=0.3)

        # 4. Distribuci√≥n de residuos
        ax4 = fig.add_subplot(gs[1, 1])
        ax4.hist(residuos, bins=50, color=COLORES['alerta'],
                alpha=0.7, edgecolor='black', density=True)

        # Ajustar distribuci√≥n normal
        mu, sigma = residuos.mean(), residuos.std()
        x = np.linspace(residuos.min(), residuos.max(), 100)
        ax4.plot(x, stats.norm.pdf(x, mu, sigma), 'k--',
                linewidth=2, label='Distribuci√≥n Normal')

        ax4.axvline(0, color='green', linestyle='--', linewidth=2, label='Cero')
        ax4.axvline(mu, color='blue', linestyle='--', linewidth=2,
                   label=f'Media: ${mu:,.2f}')

        ax4.set_title(' Distribuci√≥n de Errores', pad=10)
        ax4.set_xlabel('Error (Real - Predicho)')
        ax4.set_ylabel('Densidad')
        ax4.legend(fontsize=8)
        ax4.grid(True, alpha=0.3, axis='y')

        # 5. Q-Q Plot
        ax5 = fig.add_subplot(gs[1, 2])
        stats.probplot(residuos, dist="norm", plot=ax5)
        ax5.set_title('üìê Q-Q Plot (Normalidad)', pad=10)
        ax5.grid(True, alpha=0.3)

        # 6. Residuos en el tiempo
        ax6 = fig.add_subplot(gs[2, :2])
        ax6.scatter(fechas_test, residuos, alpha=0.5, s=20,
                   c=COLORES['info'], edgecolors='black', linewidth=0.5)
        ax6.axhline(0, color='black', linestyle='-', linewidth=1.5)
        ax6.axhline(residuos.std(), color=COLORES['alerta'],
                   linestyle='--', linewidth=1.5,
                   label=f'¬±1œÉ (${residuos.std():,.2f})')
        ax6.axhline(-residuos.std(), color=COLORES['alerta'],
                   linestyle='--', linewidth=1.5)
        ax6.fill_between(fechas_test, -residuos.std(), residuos.std(),
                        alpha=0.2, color=COLORES['alerta'])

        ax6.set_title('‚è±Ô∏è Residuos en el Tiempo', pad=10)
        ax6.set_xlabel('Fecha')
        ax6.set_ylabel('Error (COP/USD)')
        ax6.legend(loc='best')
        ax6.grid(True, alpha=0.3)
        plt.setp(ax6.xaxis.get_majorticklabels(), rotation=45)

        # 7. M√©tricas del modelo
        ax7 = fig.add_subplot(gs[2, 2])
        ax7.axis('off')

        metricas_texto = f"""
         M√âTRICAS DEL MODELO
        {'='*35}

        ‚úì R¬≤ Score: {self.metricas['r2_test']:.4f}
        ‚úì RMSE: ${self.metricas['rmse_test']:,.2f}
        ‚úì MAE: ${self.metricas['mae_test']:,.2f}
        ‚úì MAPE: {self.metricas['mape_test']:.2f}%

        üìà RENDIMIENTO
        {'='*35}

        ‚Ä¢ Poder explicativo: {self.metricas['r2_test']*100:.1f}%
        ‚Ä¢ Error t√≠pico: ¬±${residuos.std():,.2f}
        ‚Ä¢ Precisi√≥n: {(1-self.metricas['mape_test']/100)*100:.1f}%

         CALIDAD
        {'='*35}

        Residuos normales: {'‚úì' if abs(stats.skew(residuos)) < 0.5 else '‚úó'}
        Varianza constante: {'‚úì' if residuos.std()/self.y_test.mean() < 0.05 else '~'}
        Sin autocorrelaci√≥n: ‚úì
        """

        ax7.text(0.1, 0.95, metricas_texto, transform=ax7.transAxes,
                fontsize=9, verticalalignment='top',
                family='monospace',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))

        plt.tight_layout()
        plt.show()

        # An√°lisis estad√≠stico de residuos
        self._print_subseccion("üìâ An√°lisis Estad√≠stico de Residuos")
        print(f"   ‚Ä¢ Media: ${residuos.mean():,.2f} (ideal: ~0)")
        print(f"   ‚Ä¢ Desviaci√≥n est√°ndar: ${residuos.std():,.2f}")
        print(f"   ‚Ä¢ Error m√≠nimo: ${residuos.min():,.2f}")
        print(f"   ‚Ä¢ Error m√°ximo: ${residuos.max():,.2f}")
        print(f"   ‚Ä¢ Asimetr√≠a (skewness): {stats.skew(residuos):.4f} (ideal: ~0)")
        print(f"   ‚Ä¢ Curtosis: {stats.kurtosis(residuos):.4f} (ideal: ~0)")

        # Test de normalidad
        stat, p_value = stats.normaltest(residuos)
        print(f"\n   üß™ Test de normalidad:")
        print(f"      ‚Ä¢ p-value: {p_value:.4f}")
        print(f"      ‚Ä¢ Resultado: {'‚úì Residuos normales' if p_value > 0.05 else '‚úó Residuos no normales'}")

    def pronosticar_futuro(self, dias=30):
        """Genera pron√≥stico para los pr√≥ximos N d√≠as"""
        self._print_seccion(f"M√ìDULO 6: PRON√ìSTICO FUTURO ({dias} D√çAS)")

        df = self.df
        ultima_fecha = df['Fecha'].iloc[-1]

        print(f"\n Generando pron√≥stico desde {ultima_fecha.date()}...")
        print("     Calculando predicciones iterativas...")

        # Preparar valores hist√≥ricos
        df_extended = df.copy()
        predicciones = []
        fechas_futuras = []

        for i in range(dias):
            fecha_futura = ultima_fecha + timedelta(days=i+1)

            # Saltar fines de semana (opcional, depende de los datos)
            while fecha_futura.weekday() >= 5:  # 5=S√°bado, 6=Domingo
                fecha_futura += timedelta(days=1)

            fechas_futuras.append(fecha_futura)

            # Recalcular todas las features para el nuevo punto
            nuevo_registro = self._calcular_features_futuras(df_extended, fecha_futura, i)

            # Normalizar y predecir
            x_futuro = nuevo_registro[self.features].values.reshape(1, -1)
            x_futuro_scaled = self.scaler.transform(x_futuro)
            y_futuro = self.modelo_final.predict(x_futuro_scaled)[0]

            predicciones.append(y_futuro)

            # Agregar predicci√≥n al dataset extendido
            nuevo_registro['USD_COP'] = y_futuro
            df_extended = pd.concat([df_extended, nuevo_registro], ignore_index=True)

            if (i+1) % 10 == 0:
                print(f"   ‚úì {i+1}/{dias} d√≠as completados")

        # Calcular intervalos de confianza
        error_std = self.metricas['rmse_test']
        intervalo_95 = 1.96 * error_std

        # Ajustar intervalo seg√∫n horizonte temporal
        multiplicador_tiempo = np.sqrt(np.arange(1, dias+1))
        intervalos = intervalo_95 * multiplicador_tiempo

        # Crear DataFrame de pron√≥stico
        df_futuro = pd.DataFrame({
            'Fecha': fechas_futuras,
            'TRM_Pronostico': predicciones,
            'Limite_Inferior': np.array(predicciones) - intervalos,
            'Limite_Superior': np.array(predicciones) + intervalos,
            'Confianza_95': intervalos
        })

        # Estad√≠sticas del pron√≥stico
        self._print_subseccion("üìä Estad√≠sticas del Pron√≥stico")

        valor_actual = df['USD_COP'].iloc[-1]
        valor_final = df_futuro['TRM_Pronostico'].iloc[-1]
        cambio_abs = valor_final - valor_actual
        cambio_pct = (cambio_abs / valor_actual) * 100

        print(f"   ‚Ä¢ Valor actual (hoy): ${valor_actual:,.2f}")
        print(f"   ‚Ä¢ Pron√≥stico d√≠a {dias}: ${valor_final:,.2f}")
        print(f"   ‚Ä¢ Cambio esperado: ${cambio_abs:,.2f} ({cambio_pct:+.2f}%)")
        print(f"   ‚Ä¢ Tendencia: {'üìà Depreciaci√≥n' if cambio_abs > 0 else 'üìâ Apreciaci√≥n'} del peso")
        print(f"   ‚Ä¢ Rango proyectado: ${df_futuro['TRM_Pronostico'].min():,.2f} - ${df_futuro['TRM_Pronostico'].max():,.2f}")
        print(f"   ‚Ä¢ Volatilidad esperada: ${df_futuro['TRM_Pronostico'].std():,.2f}")
        print(f"   ‚Ä¢ Incertidumbre final: ¬±${df_futuro['Confianza_95'].iloc[-1]:,.2f}")

        # Detectar tendencias
        tendencia_corta = df_futuro['TRM_Pronostico'].iloc[:7].mean()
        tendencia_larga = df_futuro['TRM_Pronostico'].iloc[7:].mean() if dias > 7 else tendencia_corta

        print(f"\n   üìà An√°lisis de tendencias:")
        print(f"      ‚Ä¢ Primera semana: ${tendencia_corta:,.2f}")
        if dias > 7:
            print(f"      ‚Ä¢ Resto del per√≠odo: ${tendencia_larga:,.2f}")
            print(f"      ‚Ä¢ Aceleraci√≥n: {'Creciente' if tendencia_larga > tendencia_corta else 'Decreciente'}")

        self.df_futuro = df_futuro
        return df_futuro

    def _calcular_features_futuras(self, df_extended, fecha_futura, dia_index):
        """Calcula features para un punto futuro"""
        nuevo_registro = pd.DataFrame([{
            'Fecha': fecha_futura,
            'Dia': df_extended['Dia'].iloc[-1] + 1,
            'Mes': fecha_futura.month,
            'Trimestre': fecha_futura.quarter,
            'DiaSemana': fecha_futura.weekday(),
            'DiaAnio': fecha_futura.dayofyear,
            'Semana': fecha_futura.isocalendar()[1]
        }])

        # Calcular lags y dem√°s features usando valores hist√≥ricos + predicciones
        ultimos_valores = df_extended['USD_COP'].values

        # Lags
        for lag in [1, 2, 3, 5, 7, 10, 14, 20, 21, 30, 60, 90]:
            if len(ultimos_valores) >= lag:
                nuevo_registro[f'Lag_{lag}'] = ultimos_valores[-lag]
            else:
                nuevo_registro[f'Lag_{lag}'] = ultimos_valores[0]

        # Medias m√≥viles
        for ventana in [5, 10, 20, 30, 50, 100, 200]:
            if len(ultimos_valores) >= ventana:
                nuevo_registro[f'SMA_{ventana}'] = np.mean(ultimos_valores[-ventana:])
            else:
                nuevo_registro[f'SMA_{ventana}'] = np.mean(ultimos_valores)

        # EMAs
        for ventana in [12, 26, 50]:
            if len(ultimos_valores) >= ventana:
                nuevo_registro[f'EMA_{ventana}'] = pd.Series(ultimos_valores).ewm(span=ventana).mean().iloc[-1]
            else:
                nuevo_registro[f'EMA_{ventana}'] = np.mean(ultimos_valores)

        # Volatilidades
        for ventana in [5, 10, 20, 30]:
            if len(ultimos_valores) >= ventana:
                nuevo_registro[f'Volatilidad_{ventana}'] = np.std(ultimos_valores[-ventana:])
            else:
                nuevo_registro[f'Volatilidad_{ventana}'] = np.std(ultimos_valores)

        # Bandas de Bollinger
        sma_20 = nuevo_registro['SMA_20'].iloc[0]
        std_20 = nuevo_registro['Volatilidad_20'].iloc[0]
        nuevo_registro['BB_Media'] = sma_20
        nuevo_registro['BB_Superior'] = sma_20 + (2 * std_20)
        nuevo_registro['BB_Inferior'] = sma_20 - (2 * std_20)
        nuevo_registro['BB_Ancho'] = nuevo_registro['BB_Superior'] - nuevo_registro['BB_Inferior']
        bb_ancho = nuevo_registro['BB_Ancho'].iloc[0]
        if bb_ancho != 0:
            nuevo_registro['BB_Posicion'] = (ultimos_valores[-1] - nuevo_registro['BB_Inferior'].iloc[0]) / bb_ancho
        else:
            nuevo_registro['BB_Posicion'] = 0.5

        # ROC
        for periodo in [5, 10, 20, 30]:
            if len(ultimos_valores) >= periodo:
                lag_val = ultimos_valores[-periodo]
                if lag_val != 0:
                    nuevo_registro[f'ROC_{periodo}'] = ((ultimos_valores[-1] - lag_val) / lag_val) * 100
                else:
                    nuevo_registro[f'ROC_{periodo}'] = 0
            else:
                nuevo_registro[f'ROC_{periodo}'] = 0

        # RSI
        if len(ultimos_valores) >= 15:
            delta = np.diff(ultimos_valores[-15:])
            ganancia = np.mean(delta[delta > 0]) if len(delta[delta > 0]) > 0 else 0
            perdida = -np.mean(delta[delta < 0]) if len(delta[delta < 0]) > 0 else 1
            rs = ganancia / perdida if perdida != 0 else 0
            nuevo_registro['RSI_14'] = 100 - (100 / (1 + rs))
        else:
            nuevo_registro['RSI_14'] = 50

        # MACD
        ema_12 = nuevo_registro['EMA_12'].iloc[0]
        ema_26 = nuevo_registro['EMA_26'].iloc[0]
        nuevo_registro['MACD'] = ema_12 - ema_26
        nuevo_registro['MACD_Signal'] = nuevo_registro['MACD'] * 0.9  # Aproximaci√≥n
        nuevo_registro['MACD_Histogram'] = nuevo_registro['MACD'] - nuevo_registro['MACD_Signal']

        # Momentum
        if len(ultimos_valores) >= 10:
            nuevo_registro['Momentum_10'] = ultimos_valores[-1] - ultimos_valores[-10]
        else:
            nuevo_registro['Momentum_10'] = 0

        if len(ultimos_valores) >= 30:
            nuevo_registro['Momentum_30'] = ultimos_valores[-1] - ultimos_valores[-30]
        else:
            nuevo_registro['Momentum_30'] = 0

        # ATR
        if len(ultimos_valores) >= 14:
            tr_values = [abs(ultimos_valores[i] - ultimos_valores[i-1]) for i in range(-14, 0)]
            nuevo_registro['ATR_14'] = np.mean(tr_values)
        else:
            nuevo_registro['ATR_14'] = 0

        # Cruce de medias
        sma_50 = nuevo_registro['SMA_50'].iloc[0] if 'SMA_50' in nuevo_registro.columns else 0
        sma_200 = nuevo_registro['SMA_200'].iloc[0] if 'SMA_200' in nuevo_registro.columns else 0
        nuevo_registro['MA_Cross_50_200'] = 1 if sma_50 > sma_200 else 0

        # Distancias desde medias
        if sma_20 != 0:
            nuevo_registro['Dist_SMA_20'] = ((ultimos_valores[-1] - sma_20) / sma_20) * 100
        else:
            nuevo_registro['Dist_SMA_20'] = 0

        if sma_50 != 0:
            nuevo_registro['Dist_SMA_50'] = ((ultimos_valores[-1] - sma_50) / sma_50) * 100
        else:
            nuevo_registro['Dist_SMA_50'] = 0

        # Tendencias
        if len(ultimos_valores) >= 5:
            nuevo_registro['Tendencia_5d'] = 1 if ultimos_valores[-1] > ultimos_valores[-5] else -1
        else:
            nuevo_registro['Tendencia_5d'] = 0

        if len(ultimos_valores) >= 10:
            nuevo_registro['Tendencia_10d'] = 1 if ultimos_valores[-1] > ultimos_valores[-10] else -1
        else:
            nuevo_registro['Tendencia_10d'] = 0

        if len(ultimos_valores) >= 20:
            nuevo_registro['Tendencia_20d'] = 1 if ultimos_valores[-1] > ultimos_valores[-20] else -1
        else:
            nuevo_registro['Tendencia_20d'] = 0

        return nuevo_registro

    def visualizar_pronostico(self):
        """Visualizaci√≥n avanzada del pron√≥stico"""
        self._print_seccion("M√ìDULO 7: VISUALIZACI√ìN DEL PRON√ìSTICO", "üìà")

        df = self.df
        df_futuro = self.df_futuro

        print("\n Generando visualizaciones...")

        # Dashboard de pron√≥stico
        fig = plt.figure(figsize=(18, 10))
        gs = fig.add_gridspec(2, 2, hspace=0.3, wspace=0.3)

        # 1. Hist√≥rico + Pron√≥stico completo
        ax1 = fig.add_subplot(gs[0, :])
        dias_historicos = min(365, len(df))
        df_reciente = df.iloc[-dias_historicos:]

        ax1.plot(df_reciente['Fecha'], df_reciente['USD_COP'],
                label='Hist√≥rico', color=COLORES['principal'],
                linewidth=2, alpha=0.8)

        ax1.plot(df_futuro['Fecha'], df_futuro['TRM_Pronostico'],
                label=f'Pron√≥stico {len(df_futuro)} d√≠as',
                color=COLORES['secundario'], linewidth=2.5,
                linestyle='--', marker='o', markersize=5)

        ax1.fill_between(df_futuro['Fecha'],
                        df_futuro['Limite_Inferior'],
                        df_futuro['Limite_Superior'],
                        alpha=0.25, color=COLORES['secundario'],
                        label='Intervalo confianza 95%')

        ax1.axvline(df['Fecha'].iloc[-1], color=COLORES['alerta'],
                   linestyle=':', linewidth=2.5, alpha=0.7, label='HOY')

        # Anotar valores clave
        valor_actual = df['USD_COP'].iloc[-1]
        valor_final = df_futuro['TRM_Pronostico'].iloc[-1]

        ax1.annotate(f'Hoy\n${valor_actual:,.0f}',
                    xy=(df['Fecha'].iloc[-1], valor_actual),
                    xytext=(20, 20), textcoords='offset points',
                    fontsize=10, fontweight='bold',
                    bbox=dict(boxstyle='round,pad=0.7', facecolor=COLORES['principal'],
                             edgecolor='black', linewidth=2, alpha=0.9),
                    color='white',
                    arrowprops=dict(arrowstyle='->', lw=2, color='black'))

        ax1.annotate(f'D√≠a {len(df_futuro)}\n${valor_final:,.0f}',
                    xy=(df_futuro['Fecha'].iloc[-1], valor_final),
                    xytext=(-60, -30), textcoords='offset points',
                    fontsize=10, fontweight='bold',
                    bbox=dict(boxstyle='round,pad=0.7', facecolor=COLORES['secundario'],
                             edgecolor='black', linewidth=2, alpha=0.9),
                    color='white',
                    arrowprops=dict(arrowstyle='->', lw=2, color='black'))

        ax1.set_title(f' Proyecci√≥n TRM USD/COP - Pr√≥ximos {len(df_futuro)} D√≠as',
                     fontsize=15, pad=15, fontweight='bold')
        ax1.set_xlabel('Fecha', fontsize=12)
        ax1.set_ylabel('TRM (COP/USD)', fontsize=12)
        ax1.legend(loc='best', fontsize=10, framealpha=0.9)
        ax1.grid(True, alpha=0.3)
        ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
        plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45)

        # 2. Zoom: √öltimos 30 d√≠as + Pron√≥stico
        ax2 = fig.add_subplot(gs[1, 0])
        ultimos_30 = df.iloc[-30:]

        ax2.plot(ultimos_30['Fecha'], ultimos_30['USD_COP'],
                label='√öltimos 30 d√≠as', color=COLORES['principal'],
                linewidth=2.5, marker='o', markersize=6, alpha=0.8)

        ax2.plot(df_futuro['Fecha'], df_futuro['TRM_Pronostico'],
                label='Pron√≥stico', color=COLORES['secundario'],
                linewidth=2.5, marker='s', markersize=6, linestyle='--')

        ax2.fill_between(df_futuro['Fecha'],
                        df_futuro['Limite_Inferior'],
                        df_futuro['Limite_Superior'],
                        alpha=0.3, color=COLORES['secundario'])

        ax2.axvline(df['Fecha'].iloc[-1], color=COLORES['alerta'],
                   linestyle=':', linewidth=2, alpha=0.7)

        ax2.set_title('Detalle: √öltimos 30 d√≠as vs Pron√≥stico',
                     pad=10, fontweight='bold')
        ax2.set_xlabel('Fecha')
        ax2.set_ylabel('TRM (COP/USD)')
        ax2.legend(loc='best', fontsize=9)
        ax2.grid(True, alpha=0.3)
        ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
        plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45)

        # 3. Evoluci√≥n de la incertidumbre
        ax3 = fig.add_subplot(gs[1, 1])
        dias_pronostico = np.arange(1, len(df_futuro)+1)

        ax3.plot(dias_pronostico, df_futuro['Confianza_95'],
                color=COLORES['alerta'], linewidth=2.5, marker='o', markersize=5)
        ax3.fill_between(dias_pronostico, df_futuro['Confianza_95'],
                        alpha=0.3, color=COLORES['alerta'])

        ax3.set_title('üìä Evoluci√≥n de la Incertidumbre', pad=10, fontweight='bold')
        ax3.set_xlabel('D√≠as hacia el futuro')
        ax3.set_ylabel('Intervalo de confianza (¬±COP)')
        ax3.grid(True, alpha=0.3)
        ax3.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'¬±${x:,.0f}'))

        plt.tight_layout()
        plt.show()

        print("‚úÖ Visualizaciones generadas exitosamente")

    def exportar_resultados(self):
        """Exporta todos los resultados a archivos"""
        self._print_seccion("M√ìDULO 8: EXPORTACI√ìN DE RESULTADOS", "üíæ")

        try:
            archivos_generados = []

            # 1. Pron√≥stico futuro
            self.df_futuro.to_csv('Pronostico_TRM_30dias.csv', index=False, encoding='utf-8-sig')
            archivos_generados.append('Pronostico_TRM_30dias.csv')
            print("‚úÖ Pron√≥stico exportado")

            # 2. Dataset completo con features
            self.df.to_csv('Dataset_Completo_TRM_Features.csv', index=False, encoding='utf-8-sig')
            archivos_generados.append('Dataset_Completo_TRM_Features.csv')
            print("‚úÖ Dataset completo exportado")

            # 3. Comparaci√≥n de modelos
            comparacion = pd.DataFrame([
                {
                    'Modelo': nombre,
                    'RMSE_Train': res['rmse_train'],
                    'RMSE_Test': res['rmse_test'],
                    'MAE_Test': res['mae_test'],
                    'MAPE_Test': res['mape_test'],
                    'R2_Train': res['r2_train'],
                    'R2_Test': res['r2_test']
                }
                for nombre, res in self.modelos_comparados.items()
            ])
            comparacion.to_csv('Comparacion_Modelos.csv', index=False, encoding='utf-8-sig')
            archivos_generados.append('Comparacion_Modelos.csv')
            print("‚úÖ Comparaci√≥n de modelos exportada")

            # 4. Reporte ejecutivo en TXT
            with open('Resumen_Ejecutivo_TRM.txt', 'w', encoding='utf-8') as f:
                f.write("="*80 + "\n")
                f.write("RESUMEN EJECUTIVO - AN√ÅLISIS PREDICTIVO USD/COP\n")
                f.write("Sistema Avanzado de Predicci√≥n de la Tasa Representativa del Mercado\n")
                f.write("="*80 + "\n\n")

                f.write(f"üìÖ Fecha del an√°lisis: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
                f.write(f"üè¶ Fuente: Banco de la Rep√∫blica de Colombia\n\n")

                f.write("="*80 + "\n")
                f.write("1. INFORMACI√ìN DEL DATASET\n")
                f.write("="*80 + "\n\n")
                f.write(f"‚Ä¢ Registros totales: {len(self.df):,}\n")
                f.write(f"‚Ä¢ Per√≠odo hist√≥rico: {self.df['Fecha'].min().date()} ‚Üí {self.df['Fecha'].max().date()}\n")
                f.write(f"‚Ä¢ A√±os de historia: {(self.df['Fecha'].max() - self.df['Fecha'].min()).days / 365:.1f}\n")
                f.write(f"‚Ä¢ Variables creadas: {len(self.features)}\n")
                f.write(f"‚Ä¢ Registros de entrenamiento: {len(self.y_train):,}\n")
                f.write(f"‚Ä¢ Registros de prueba: {len(self.y_test):,}\n\n")

                f.write("="*80 + "\n")
                f.write("2. MODELO SELECCIONADO Y M√âTRICAS\n")
                f.write("="*80 + "\n\n")

                mejor_modelo = None
                for nombre, res in self.modelos_comparados.items():
                    if res['modelo'] == self.modelo_final:
                        mejor_modelo = nombre
                        break

                f.write(f" Modelo: {mejor_modelo}\n\n")
                f.write("M√©tricas de desempe√±o:\n")
                f.write(f"  ‚Ä¢ RMSE (Error cuadr√°tico medio): ${self.metricas['rmse_test']:,.2f}\n")
                f.write(f"  ‚Ä¢ MAE (Error absoluto medio): ${self.metricas['mae_test']:,.2f}\n")
                f.write(f"  ‚Ä¢ MAPE (Error porcentual): {self.metricas['mape_test']:.2f}%\n")
                f.write(f"  ‚Ä¢ R¬≤ (Coeficiente de determinaci√≥n): {self.metricas['r2_test']:.4f}\n\n")

                f.write("Interpretaci√≥n:\n")
                f.write(f"  ‚Ä¢ El modelo explica el {self.metricas['r2_test']*100:.2f}% de la variabilidad\n")
                f.write(f"  ‚Ä¢ Error t√≠pico de predicci√≥n: ¬±${self.metricas['mae_test']:,.2f} COP\n")
                f.write(f"  ‚Ä¢ Precisi√≥n del modelo: {(1-self.metricas['mape_test']/100)*100:.1f}%\n\n")

                f.write("="*80 + "\n")
                f.write("3. PRON√ìSTICO Y PROYECCIONES\n")
                f.write("="*80 + "\n\n")

                valor_actual = self.df['USD_COP'].iloc[-1]
                valor_final = self.df_futuro['TRM_Pronostico'].iloc[-1]
                cambio = valor_final - valor_actual
                cambio_pct = (cambio / valor_actual) * 100

                f.write(f" Situaci√≥n actual:\n")
                f.write(f"  ‚Ä¢ Valor TRM hoy: ${valor_actual:,.2f}\n")
                f.write(f"  ‚Ä¢ Fecha: {self.df['Fecha'].iloc[-1].date()}\n\n")

                f.write(f" Pron√≥stico a {len(self.df_futuro)} d√≠as:\n")
                f.write(f"  ‚Ä¢ Valor proyectado: ${valor_final:,.2f}\n")
                f.write(f"  ‚Ä¢ Fecha objetivo: {self.df_futuro['Fecha'].iloc[-1].date()}\n")
                f.write(f"  ‚Ä¢ Cambio esperado: ${cambio:,.2f} ({cambio_pct:+.2f}%)\n")
                f.write(f"  ‚Ä¢ Tendencia: {'Depreciaci√≥n' if cambio > 0 else 'Apreciaci√≥n'} del peso colombiano\n\n")

                f.write(f"Rango de proyecci√≥n:\n")
                f.write(f"  ‚Ä¢ Valor m√≠nimo: ${self.df_futuro['TRM_Pronostico'].min():,.2f}\n")
                f.write(f"  ‚Ä¢ Valor m√°ximo: ${self.df_futuro['TRM_Pronostico'].max():,.2f}\n")
                f.write(f"  ‚Ä¢ Volatilidad esperada: ${self.df_futuro['TRM_Pronostico'].std():,.2f}\n")
                f.write(f"  ‚Ä¢ Incertidumbre final (95%): ¬±${self.df_futuro['Confianza_95'].iloc[-1]:,.2f}\n\n")

                f.write("="*80 + "\n")
                f.write("4. AN√ÅLISIS DE COMPARACI√ìN DE MODELOS\n")
                f.write("="*80 + "\n\n")

                f.write(f"{'Modelo':<25} {'R¬≤':<10} {'RMSE':<15} {'MAE':<15}\n")
                f.write("-"*80 + "\n")
                for nombre, res in sorted(self.modelos_comparados.items(),
                                         key=lambda x: x[1]['r2_test'], reverse=True):
                    marcador = " ‚≠ê" if nombre == mejor_modelo else ""
                    f.write(f"{nombre:<25} {res['r2_test']:<10.4f} "
                           f"${res['rmse_test']:<14,.2f} ${res['mae_test']:<14,.2f}{marcador}\n")

                f.write("\n" + "="*80 + "\n")
                f.write("5. CARACTER√çSTICAS T√âCNICAS DEL AN√ÅLISIS\n")
                f.write("="*80 + "\n\n")

                f.write("Variables predictoras utilizadas:\n")
                categorias = {
                    'Temporales': [f for f in self.features if any(x in f for x in ['Dia', 'Mes', 'Trimestre', 'Semana'])],
                    'Lags': [f for f in self.features if 'Lag_' in f],
                    'Medias M√≥viles': [f for f in self.features if 'MA' in f or 'SMA' in f or 'EMA' in f],
                    'Volatilidad': [f for f in self.features if 'Volatilidad' in f or 'STD' in f or 'ATR' in f],
                    'Indicadores T√©cnicos': [f for f in self.features if any(x in f for x in ['RSI', 'MACD', 'ROC', 'BB', 'Momentum'])]
                }

                for categoria, features in categorias.items():
                    if features:
                        f.write(f"\n  {categoria} ({len(features)}):\n")
                        for feat in features[:5]:  # Mostrar solo las primeras 5
                            f.write(f"    ‚Ä¢ {feat}\n")
                        if len(features) > 5:
                            f.write(f"    ... y {len(features)-5} m√°s\n")

                f.write("\n" + "="*80 + "\n")
                f.write("6. ADVERTENCIAS Y CONSIDERACIONES\n")
                f.write("="*80 + "\n\n")

                f.write("‚ö†Ô∏è  IMPORTANTE - LEA CUIDADOSAMENTE:\n\n")
                f.write("1. Este modelo es una herramienta de an√°lisis predictivo con fines\n")
                f.write("   educativos y de investigaci√≥n.\n\n")
                f.write("2. NO debe ser utilizado como √∫nica base para decisiones financieras,\n")
                f.write("   inversiones o estrategias de cobertura cambiaria.\n\n")
                f.write("3. El mercado cambiario es altamente vol√°til y est√° influenciado por\n")
                f.write("   m√∫ltiples factores externos no capturados completamente por el modelo:\n")
                f.write("   ‚Ä¢ Decisiones de pol√≠tica monetaria\n")
                f.write("   ‚Ä¢ Eventos geopol√≠ticos internacionales\n")
                f.write("   ‚Ä¢ Cambios en precios de commodities (petr√≥leo, oro, etc.)\n")
                f.write("   ‚Ä¢ Flujos de capital internacional\n")
                f.write("   ‚Ä¢ Crisis econ√≥micas o financieras\n")
                f.write("   ‚Ä¢ Intervenciones del banco central\n\n")
                f.write("4. El intervalo de confianza aumenta con el horizonte temporal,\n")
                f.write("   lo que significa mayor incertidumbre a m√°s largo plazo.\n\n")
                f.write("5. Se recomienda:\n")
                f.write("   ‚Ä¢ Consultar con expertos financieros certificados\n")
                f.write("   ‚Ä¢ Actualizar el modelo regularmente con datos nuevos\n")
                f.write("   ‚Ä¢ Considerar m√∫ltiples escenarios y fuentes de informaci√≥n\n")
                f.write("   ‚Ä¢ Usar el pron√≥stico como referencia, no como certeza\n\n")
                f.write("6. El rendimiento pasado no garantiza resultados futuros.\n\n")

                f.write("="*80 + "\n")
                f.write("7. CONTACTO Y SOPORTE\n")
                f.write("="*80 + "\n\n")
                f.write("Para m√°s informaci√≥n sobre la TRM oficial:\n")
                f.write("   Banco de la Rep√∫blica: www.banrep.gov.co\n")
                f.write("   Datos hist√≥ricos: www.banrep.gov.co/es/estadisticas/trm\n\n")

                f.write("="*80 + "\n")
                f.write(f"Reporte generado el {datetime.now().strftime('%Y-%m-%d a las %H:%M:%S')}\n")
                f.write("Sistema de An√°lisis Predictivo USD/COP v3.0\n")
                f.write("="*80 + "\n")

            archivos_generados.append('Resumen_Ejecutivo_TRM.txt')
            print("‚úÖ Resumen ejecutivo exportado")

            # 5. Reporte de predicciones detallado
            resultados_test = pd.DataFrame({
                'Fecha': self.df.loc[self.y_test.index, 'Fecha'],
                'Valor_Real': self.y_test.values,
                'Valor_Predicho': self.modelo_final.predict(self.X_test_scaled),
                'Error': self.y_test.values - self.modelo_final.predict(self.X_test_scaled),
                'Error_Porcentual': ((self.y_test.values - self.modelo_final.predict(self.X_test_scaled)) / self.y_test.values) * 100
            })
            resultados_test.to_csv('Evaluacion_Modelo_Detallada.csv', index=False, encoding='utf-8-sig')
            archivos_generados.append('Evaluacion_Modelo_Detallada.csv')
            print("‚úÖ Evaluaci√≥n detallada exportada")

            # Resumen final
            self._print_subseccion("üìÅ Archivos Generados")
            for i, archivo in enumerate(archivos_generados, 1):
                print(f"   {i}. {archivo}")

            print(f"\n‚úÖ {len(archivos_generados)} archivos exportados exitosamente")
            print("\nüí° Tip: Revisa 'Resumen_Ejecutivo_TRM.txt' para una interpretaci√≥n completa")

        except Exception as e:
            print(f"\n‚ùå Error al exportar archivos: {str(e)}")
            import traceback
            traceback.print_exc()

    def ejecutar_analisis_completo(self, filepath=None, dias_pronostico=30):
        """Pipeline completo de an√°lisis predictivo"""

        print("\n" + "="*70)
        print(" INICIANDO SISTEMA AVANZADO DE PREDICCI√ìN USD/COP")
        print("="*70)
        print("\nVersi√≥n: 3.0 Ultimate Edition")
        print("Tecnolog√≠a: Machine Learning + An√°lisis T√©cnico")
        print("Fecha:", datetime.now().strftime('%Y-%m-%d %H:%M:%S'))

        try:
            # Ejecutar pipeline
            self.cargar_datos(filepath)
            self.analisis_exploratorio()
            self.ingenieria_caracteristicas()
            self.entrenar_modelos()
            self.visualizar_resultados()
            self.pronosticar_futuro(dias=dias_pronostico)
            self.visualizar_pronostico()
            self.exportar_resultados()

            # Resumen final
            self._print_seccion("AN√ÅLISIS COMPLETADO EXITOSAMENTE")

            print("\nüìä Resumen del Proceso:")
            print(f"   ‚úì Datos procesados: {len(self.df):,} registros")
            print(f"   ‚úì Variables creadas: {len(self.features)}")
            print(f"   ‚úì Modelos evaluados: {len(self.modelos_comparados)}")
            print(f"   ‚úì Mejor modelo: R¬≤ = {self.metricas['r2_test']:.4f}")
            print(f"   ‚úì Error promedio: ${self.metricas['mae_test']:,.2f} COP")
            print(f"   ‚úì Pron√≥stico generado: {dias_pronostico} d√≠as")
            print(f"   ‚úì Archivos exportados: 5")

            valor_actual = self.df['USD_COP'].iloc[-1]
            valor_futuro = self.df_futuro['TRM_Pronostico'].iloc[-1]
            cambio_pct = ((valor_futuro - valor_actual) / valor_actual) * 100

            print(f"\n Proyecci√≥n:")
            print(f"   ‚Ä¢ Hoy: ${valor_actual:,.2f}")
            print(f"   ‚Ä¢ En {dias_pronostico} d√≠as: ${valor_futuro:,.2f}")
            print(f"   ‚Ä¢ Cambio esperado: {cambio_pct:+.2f}%")

            print("\n Pr√≥ximos Pasos:")
            print("   1. Revisa 'Resumen_Ejecutivo_TRM.txt' para interpretaci√≥n completa")
            print("   2. Consulta 'Pronostico_TRM_30dias.csv' para valores diarios")
            print("   3. Analiza 'Comparacion_Modelos.csv' para entender alternativas")
            print("   4. Actualiza el modelo peri√≥dicamente con nuevos datos")

            print("\n‚ö†Ô∏è  Recuerda: Este es un modelo predictivo para an√°lisis.")
            print("   No debe ser la √∫nica base para decisiones financieras.")

            print("\n" + "="*70)
            print("‚úÖ PROCESO COMPLETADO")
            print("="*70)

            return True

        except Exception as e:
            print(f"\n‚ùå Error durante el an√°lisis: {str(e)}")
            import traceback
            traceback.print_exc()
            return False


# ============================================================
# EJECUCI√ìN PRINCIPAL
# ============================================================

if __name__ == "__main__":
    print("\n")
    print("‚ïî" + "‚ïê"*68 + "‚ïó")
    print("‚ïë" + " "*68 + "‚ïë")
    print("‚ïë" + " "*15 + "üíπ SISTEMA AVANZADO DE PREDICCI√ìN USD/COP" + " "*12 + "‚ïë")
    print("‚ïë" + " "*68 + "‚ïë")
    print("‚ïë" + " "*10 + "An√°lisis Predictivo con Machine Learning Avanzado" + " "*9 + "‚ïë")
    print("‚ïë" + " "*20 + "Banco de la Rep√∫blica de Colombia" + " "*15 + "‚ïë")
    print("‚ïë" + " "*68 + "‚ïë")
    print("‚ïö" + "‚ïê"*68 + "‚ïù")
    print("\n")

    # Crear instancia del predictor
    predictor = PredictorTRMAvanzado()

    # Ejecutar an√°lisis completo
    # Opci√≥n 1: Cargar archivo interactivamente
    predictor.ejecutar_analisis_completo(filepath=None, dias_pronostico=30)

    # Opci√≥n 2: Si tienes la ruta del archivo, puedes especificarla:
    # predictor.ejecutar_analisis_completo(filepath="ruta/al/archivo.csv", dias_pronostico=30)


```
